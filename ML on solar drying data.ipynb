```python
import pandas as pd
```

```python
df_guava = pd.read_csv('/content/drive/MyDrive/GuavaData.csv')
```

```python
df_beetroot = pd.read_csv('/content/drive/MyDrive/BeetrootData.csv')
```

```python
df_muskmelon = pd.read_csv('/content/drive/MyDrive/MuskmelonData.csv')
```

```python
X_guava = df_guava.drop('moisture ratio', axis='columns')

y_guava = df_guava['moisture ratio']

X_beetroot = df_beetroot.drop('moisture ratio', axis='columns')

y_beetroot = df_beetroot['moisture ratio']

X_muskmelon = df_muskmelon.drop('moisture ratio', axis='columns')

y_muskmelon = df_muskmelon['moisture ratio']
```

```python
from sklearn.model_selection import train_test_split

X_train_guava, X_test_guava, y_train_guava, y_test_guava = train_test_split(X_guava,y_guava,test_size=0.2)
```

```python
X_train_beetroot, X_test_beetroot, y_train_beetroot, y_test_beetroot = train_test_split(X_beetroot,y_beetroot,test_size=0.2) 

X_train_muskmelon, X_test_muskmelon, y_train_muskmelon, y_test_muskmelon = train_test_split(X_muskmelon,y_muskmelon,test_size=0.2) 
```

```python
X_train_guava.insert(2, 'food sample', 'guava')
```

```python
X_train_beetroot.insert(2, 'food sample', 'beetroot')
X_train_muskmelon.insert(2, 'food sample', 'muskmelon')
```

```python
X_train_final = pd.DataFrame()
X_train_final = pd.concat([X_train_guava,X_train_beetroot,X_train_muskmelon])
X_train_final
```

```python
y_train_final = pd.DataFrame()
y_train_final = pd.concat([y_train_guava,y_train_beetroot,y_train_muskmelon])
y_train_final
```

```python
dummies = pd.get_dummies(X_train_final['food sample'])
merged = pd.concat([X_train_final,dummies],axis='columns')
X_train_final_ohe = merged.drop(['food sample','muskmelon'],axis='columns')
```

```python
X_train_final_ohe
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_final_ohe,y_train_final)
```

```python

X_test_guava.insert(2, 'beetroot', '0')
X_test_guava.insert(3, 'guava', '1')
y_pred_guava=model.predict(X_test_guava)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
lr_MSE_guava=mean_squared_error(y_test_guava, y_pred_guava)
lr_MAE_guava= mean_absolute_error(y_test_guava, y_pred_guava)
lr_RMSE_guava = np.sqrt(mean_squared_error(y_test_guava, y_pred_guava))
lr_R2_guava=r2_score(y_test_guava, y_pred_guava)
```

```python
results = pd.DataFrame({'Algorithm':['Linear Regression Method'], 'Food Sample':['Guava'],'R2':[lr_R2_guava] , 'MAE':[lr_MAE_guava], 'MSE':[lr_MSE_guava], 'RMSE':[lr_RMSE_guava] })
results
```

```python
y_pred_beetroot=model.predict(X_test_beetroot)
lr_MSE_beetroot=mean_squared_error(y_test_beetroot, y_pred_beetroot)
lr_MAE_beetroot= mean_absolute_error(y_test_beetroot, y_pred_beetroot)
lr_RMSE_beetroot = np.sqrt(mean_squared_error(y_test_beetroot, y_pred_beetroot))
lr_R2_beetroot=r2_score(y_test_beetroot, y_pred_beetroot)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Linear Regression Method'], 'Food Sample':['beetroot'],'R2':[lr_R2_beetroot] , 'MAE':[lr_MAE_beetroot], 'MSE':[lr_MSE_beetroot], 'RMSE':[lr_RMSE_beetroot] })

results = pd.concat([results, tempResults])

results
```

```python
y_pred_muskmelon=model.predict(X_test_muskmelon)
lr_MSE_muskmelon=mean_squared_error(y_test_muskmelon, y_pred_muskmelon)
lr_MAE_muskmelon= mean_absolute_error(y_test_muskmelon, y_pred_muskmelon)
lr_RMSE_muskmelon = np.sqrt(mean_squared_error(y_test_muskmelon, y_pred_muskmelon))
lr_R2_muskmelon=r2_score(y_test_muskmelon, y_pred_muskmelon)

tempResults = pd.DataFrame({'Algorithm':['Linear Regression Method'], 'Food Sample':['muskmelon'],'R2':[lr_R2_muskmelon] , 'MAE':[lr_MAE_muskmelon], 'MSE':[lr_MSE_muskmelon], 'RMSE':[lr_RMSE_muskmelon] })

results = pd.concat([results, tempResults])

results
```
