```
import pandas as pd
df = pd.read_csv('data.csv')
df
```

```
dummies = pd.get_dummies(df['food sample'])
dummies
```

```
merged = pd.concat([df,dummies],axis='columns')
```

```
final = merged.drop(['food sample','muskmelon'],axis='columns')
final
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
X = final.drop('moisture ratio', axis='columns')
X
```

```
y=final['moisture ratio']
```

```
model.fit()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) 
```

```
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
model.score(X_test, y_test)
lr_MSE=mean_squared_error(y_test, y_pred)
lr_MAE= mean_absolute_error(y_test, y_pred)
lr_RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
lr_R2=r2_score(y_test, y_pred)
```

```
results = pd.DataFrame({'Algorithm':['Linear Regression Method'],'R2':[lr_R2] , 'MAE':[lr_MAE], 'MSE':[lr_MSE], 'RMSE':[lr_RMSE] })
results
```
